## Transform

### AgeSex

```python
ZCTA_AgeSex['TotalPopulation'] = ZCTA_AgeSex['B01001_001']
ZCTA_AgeSex['FemalePopulation'] = ZCTA_AgeSex['B01001_026']/ZCTA_AgeSex['B01001_001']
ZCTA_AgeSex['MalePopulation'] = ZCTA_AgeSex['B01001_002']/ZCTA_AgeSex['B01001_001']
ZCTA_AgeSex = ZCTA_AgeSex[['ZCTA', 'TotalPopulation', 'FemalePopulation', 'MalePopulation']]
```

```python
ZCTA = ZCTA_AgeSex
ZCTA.head()
```

### HealthInsurance

```python
ZCTA_HealthInsurance.head()
```

```python
ZCTA_HealthInsurance['NoninstitutionalizedCivilians'] = ZCTA_HealthInsurance['B27001_001']
ZCTA_HealthInsurance['Insured'] = ZCTA_HealthInsurance['Insured'] / ZCTA_HealthInsurance['NoninstitutionalizedCivilians']
ZCTA_HealthInsurance['Uninsured'] = ZCTA_HealthInsurance['Uninsured'] / ZCTA_HealthInsurance['NoninstitutionalizedCivilians']
ZCTA_HealthInsurance = ZCTA_HealthInsurance[['ZCTA', 'Insured', 'Uninsured']]
ZCTA_HealthInsurance.head()
```

```python
ZCTA = pd.merge(ZCTA_HealthInsurance, ZCTA_AgeSex, on="ZCTA")
ZCTA
```

### Race

```python
ZCTA_Race.head()
```

```python
ZCTA_Race['White'] = ZCTA_Race['B02008_001'] / ZCTA_Race['C02003_001']
ZCTA_Race['AfricanAmerican'] = ZCTA_Race['B02009_001'] / ZCTA_Race['C02003_001']
ZCTA_Race['NativeAmerican'] = ZCTA_Race['B02010_001'] / ZCTA_Race['C02003_001']
ZCTA_Race['Asian'] = ZCTA_Race['B02011_001'] / ZCTA_Race['C02003_001']
ZCTA_Race['PacificIslander'] = ZCTA_Race['B02012_001'] / ZCTA_Race['C02003_001']
ZCTA_Race = ZCTA_Race[['ZCTA', 'White', 'AfricanAmerican', 'NativeAmerican', 'Asian', 'PacificIslander']]
ZCTA_Race.head()
```

```python
ZCTA = pd.merge(ZCTA, ZCTA_Race, on="ZCTA")
ZCTA.head()
```

### HispanicOrigin

```python
ZCTA_HispanicOrigin.head()
```

```python
ZCTA_HispanicOrigin['Hispanic'] = ZCTA_HispanicOrigin['B03001_003'] / ZCTA_HispanicOrigin['B03001_001']
ZCTA_HispanicOrigin = ZCTA_HispanicOrigin[['ZCTA', 'Hispanic']]
ZCTA_HispanicOrigin.head()
```

```python
ZCTA = pd.merge(ZCTA, ZCTA_HispanicOrigin, on="ZCTA")
ZCTA.head()
```

### ForeignBirth

```python
ZCTA_ForeignBirth.head()
```

```python
ZCTA = pd.merge(ZCTA, ZCTA_ForeignBirth, on="ZCTA")
ZCTA.head()
```

### Grandparents

```python
ZCTA_Grandparents.head()
```

```python
ZCTA_Grandparents['GrandparentHouseholder'] = ZCTA_Grandparents['B10001_001'] / ZCTA['TotalPopulation']
ZCTA_Grandparents['GrandparentPresence'] = ZCTA_Grandparents['B10051_001'] / ZCTA['TotalPopulation']
ZCTA_Grandparents = ZCTA_Grandparents[['ZCTA', 'GrandparentHouseholder', 'GrandparentPresence']]
ZCTA_Grandparents.head()
```

```python
ZCTA = pd.merge(ZCTA, ZCTA_Grandparents, on="ZCTA")
ZCTA.head()
```

### HouseholdsFamilies

```python
ZCTA_HouseholdsFamilies.head()
```

```python
ZCTA_HouseholdsFamilies['FamilyHousehold'] = ZCTA_HouseholdsFamilies['B11016_002'] / ZCTA_HouseholdsFamilies['B11016_001']
ZCTA_HouseholdsFamilies['NonfamilyHousehold'] = ZCTA_HouseholdsFamilies['B11016_009'] / ZCTA_HouseholdsFamilies['B11016_001']
ZCTA_HouseholdsFamilies = ZCTA_HouseholdsFamilies[['ZCTA', 'FamilyHousehold', 'NonfamilyHousehold']]
ZCTA_HouseholdsFamilies.head()
```

```python
ZCTA = pd.merge(ZCTA, ZCTA_HouseholdsFamilies, on="ZCTA")
ZCTA.head()
```

### Income

```python
ZCTA_Income.head()
```

```python
ZCTA_Income["B19013_001"].replace({'.': '0'}, inplace=True)
ZCTA_Income['MedianHouseholdIncome'] = pd.to_numeric(ZCTA_Income['B19013_001'])
ZCTA_Income = ZCTA_Income[['ZCTA', 'MedianHouseholdIncome']]
ZCTA = pd.merge(ZCTA, ZCTA_Income, on="ZCTA")
ZCTA.head()
```

### JourneyToWork

```python
ZCTA_JourneyToWork.head()
```

```python
ZCTA_JourneyToWork['CarTruckVan'] = ZCTA_JourneyToWork['B08006_002'] / ZCTA_JourneyToWork['B08006_001']
ZCTA_JourneyToWork['PublicTransport'] = ZCTA_JourneyToWork['B08006_008'] / ZCTA_JourneyToWork['B08006_001']
ZCTA_JourneyToWork['Bicycle'] = ZCTA_JourneyToWork['B08006_014'] / ZCTA_JourneyToWork['B08006_001']
ZCTA_JourneyToWork['Walked'] = ZCTA_JourneyToWork['B08006_015'] / ZCTA_JourneyToWork['B08006_001']
ZCTA_JourneyToWork['WFH'] = ZCTA_JourneyToWork['B08006_017'] / ZCTA_JourneyToWork['B08006_001']
ZCTA_JourneyToWork = ZCTA_JourneyToWork[['ZCTA', 'CarTruckVan', 'PublicTransport', 'Bicycle', 'Walked', 'WFH']]
ZCTA = pd.merge(ZCTA, ZCTA_JourneyToWork, on="ZCTA")
ZCTA.head()
```

### Poverty

```python
ZCTA_Poverty.head()
```

```python
ZCTA_Poverty['BelowPoverty'] = ZCTA_Poverty['B17001_002'] / ZCTA_Poverty['B17001_001']
ZCTA_Poverty = ZCTA_Poverty[['ZCTA', 'BelowPoverty']]
ZCTA = pd.merge(ZCTA, ZCTA_Poverty, on="ZCTA")
ZCTA.head()
```

### Earnings

```python
ZCTA_Earnings.head()
```

```python
ZCTA_Earnings = ZCTA_Earnings.rename(columns={'B20002_001':'MedianEarnings', 'B20002_002':'MedianEarningsMale', 'B20002_003':'MedianEarningsFemale'})
ZCTA_Earnings
```

```python
ZCTA_Earnings["MedianEarnings"].replace({'.': '0'}, inplace=True)
ZCTA_Earnings['MedianEarnings'] = pd.to_numeric(ZCTA_Earnings['MedianEarnings'])

ZCTA_Earnings["MedianEarningsMale"].replace({'.': '0'}, inplace=True)
ZCTA_Earnings['MedianEarningsMale'] = pd.to_numeric(ZCTA_Earnings['MedianEarningsMale'])

ZCTA_Earnings["MedianEarningsFemale"].replace({'.': '0'}, inplace=True)
ZCTA_Earnings['MedianEarningsFemale'] = pd.to_numeric(ZCTA_Earnings['MedianEarningsFemale'])
```

```python
ZCTA_Earnings['MedianEarningsGenderInequality'] = (ZCTA_Earnings['MedianEarningsMale'] - ZCTA_Earnings['MedianEarningsFemale']) / ZCTA_Earnings['MedianEarningsFemale']
ZCTA = pd.merge(ZCTA, ZCTA_Earnings, on="ZCTA")
ZCTA.head()
```

### SchoolEnrollment

```python
ZCTA_SchoolEnrollment.head()
```

```python
ZCTA_SchoolEnrollment = ZCTA_SchoolEnrollment.rename(columns={'B14001_007':'EnrolledInHighschool', 'B14001_008':'EnrolledInUndergrad', 'B14001_009':'EnrolledInGrad'})
ZCTA_SchoolEnrollment['EnrolledInSchool'] = ZCTA_SchoolEnrollment['B14001_003'] + ZCTA_SchoolEnrollment['B14001_004'] + ZCTA_SchoolEnrollment['B14001_005'] + ZCTA_SchoolEnrollment['B14001_006']
ZCTA_SchoolEnrollment['EnrolledInSchool'] = ZCTA_SchoolEnrollment['EnrolledInSchool'] / ZCTA_SchoolEnrollment['B14001_001']
ZCTA_SchoolEnrollment['EnrolledInHighschool'] = ZCTA_SchoolEnrollment['EnrolledInHighschool'] / ZCTA_SchoolEnrollment['B14001_001']
ZCTA_SchoolEnrollment['EnrolledInUndergrad'] = ZCTA_SchoolEnrollment['EnrolledInUndergrad'] / ZCTA_SchoolEnrollment['B14001_001']
ZCTA_SchoolEnrollment['EnrolledInGrad'] = ZCTA_SchoolEnrollment['EnrolledInGrad'] / ZCTA_SchoolEnrollment['B14001_001']

ZCTA_SchoolEnrollment = ZCTA_SchoolEnrollment[['ZCTA', 'EnrolledInSchool', 'EnrolledInHighschool', 'EnrolledInUndergrad', 'EnrolledInGrad']]
ZCTA = pd.merge(ZCTA, ZCTA_SchoolEnrollment, on="ZCTA")
ZCTA.head()
```

### EmploymentStatus

```python
ZCTA_EmploymentStatus.head()
```

```python
ZCTA_EmploymentStatus = ZCTA_EmploymentStatus.rename(columns={'B23025_004':'Employed', 'B23025_005':'Unemployed'})
ZCTA_EmploymentStatus['Employed'] = ZCTA_EmploymentStatus['Employed'] / ZCTA_EmploymentStatus['B23025_003']
ZCTA_EmploymentStatus['Unemployed'] = ZCTA_EmploymentStatus['Unemployed'] / ZCTA_EmploymentStatus['B23025_003']

ZCTA_EmploymentStatus = ZCTA_EmploymentStatus[['ZCTA', 'Employed', 'Unemployed']]
ZCTA = pd.merge(ZCTA, ZCTA_EmploymentStatus, on="ZCTA")
ZCTA.head()
```
